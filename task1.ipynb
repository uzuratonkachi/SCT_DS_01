```python
cols_to_keep = ['Year', 'Population']
df = pd.read_csv(r"C:\Users\megha\Downloads\India_Population.csv", usecols= cols_to_keep)
df1 = df.set_index("Year")
df1
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
df1.plot(kind="bar" )
plt.xticks(rotation=45)
plt.ylabel("Population")
plt.gca().invert_xaxis()
plt.show()
```
